**Why were these datasets chosen?** 

The **COVID-19** pandemic, also famous as the coronavirus pandemic now a days, is an ongoing epidemic of coronavirus disease 2019 (COVID 19). The outbreak was identified in Wuhan, China, in December 2019. And within a month the ***“World Health Organization”*** declared a Public Health Emergency of International Concern on 30 January, and a pandemic on 11 March.  The virus is primarily spread between people in close contact, often via small droplets produced by coughing, sneezing, and talking. As of now, more than 3 million cases of COVID-19 have been reported in over 187 countries and territories, resulting in more than 274,000 deaths. This ranges from mandatory geographic quarantines to non-mandatory recommendations to stay at home. 

***Response to the pandemic:***

It is interesting to see how Scandinavian countries manage to contain the virus spread. It is relevant to roughly look into the different strategies of the Scandinavian countries. The populations of **Denmark, Norway** and **Finland** are similar, and they have chosen very similar strategies after their first cases at the end of February. The first case of COVID-19 was reported in **Denmark** and **Norway** in the late February. Within the next two weeks, countries announces a national lockdown also imposing a travel ban and closing the borders. However, Finland had noticed the first case of corona at the end of January, but managed to contain the spread at the time by quarantine and contact tracing. A month later, on February 26, Finland identified another case of corona and the spread. Hence, Finland declared a state of emergency on March 16, closed schools and public facilities and restrcited the national borders[(1)](https://www.diis.dk/en/research/the-nordics-and-covid-19-together-or-apart).

On the other hand, **Sweden** with roughly double inhabitants of the other countries, has chosen a very different strategy. They have maintained society almost the same way as before its first case in late January. They have borders open for other members of the European Union, and schools, restaurants, shops and bars remain open. **Iceland** has a small population with around 365 000 people. However, it also has an infection rate of 1 case per 204 inhabitants, one of the highest in world. This could be attributed to more tests carried out per capita than any other country[(1)](https://www.diis.dk/en/research/the-nordics-and-covid-19-together-or-apart).

## Visualisation 3
> Population density : cases per thousand  

We take a look at the different trends of virus growth in each country. Further want to see an indication of other trends upholding factors such as population density. The below illustration shows the map of Scandinavia. You can get an overview of the map which shows population density by confirmed cases per 1000 for each country with a darker or lighter shade.





```
def popdens(row):
  if row['Code'] == 'DNK':
    return row['5/13/20'] / pop.at[0, 'Population'] * 1000
  if row['Code'] == 'FIN':
    return row['5/13/20'] / pop.at[1, 'Population'] * 1000
  if row['Code'] == 'ISL':
    return row['5/13/20'] / pop.at[2, 'Population'] * 1000
  if row['Code'] == 'NOR':
    return row['5/13/20'] / pop.at[3, 'Population'] * 1000
  if row['Code'] == 'SWE':
    return row['5/13/20'] / pop.at[4, 'Population'] * 1000
  else:
    return "error"
```

```
pop = pd.read_csv('https://raw.githubusercontent.com/eng22/misc/master/Populations.csv')
data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
my_countries = ['Denmark', 'Finland', 'Iceland', 'Norway', 'Sweden']
my_data = data.loc[data['Country/Region'].isin(my_countries)]
#Group the data by the scandinavian country 
my_data = my_data.groupby('Country/Region').sum()
my_data['Code'] = ['DNK', 'FIN', 'ISL', 'NOR', 'SWE' ]
my_data['CasesPerThou'] = my_data.apply (lambda row: popdens(row), axis=1)

fig = go.Figure(data=go.Choropleth(
    locations = my_data['Code'],
    locationmode='ISO-3',
    z = my_data['CasesPerThou'],
    text = my_data.loc[:, '5/13/20'],
    colorscale = 'darkmint',
    autocolorscale=True,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.004,
    colorbar_tickprefix = '',
    colorbar_title = 'Cases per 1000',
))

fig.update_layout(
    title_text='Cases per Thousand',
    geo=dict(
        scope='europe', showframe=False,
        showcoastlines=False,
        projection_type='equirectangular'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='',
        showarrow = False
    )]
)

fig.show()
```

Looking at the map and hovering the cursor will show the number of COVID-19 cases with respect to the population of each country. Here, the lighter shade addresses the less number of confirmed cases per 1000 occupants of the country. Where Finland has the lowest ratio of 1.07 followed by Norway and Denmark, 1.48 and 1.88 respectively. In contrast, Sweden is the second highest at 2.75. Interestingly, Iceland has the highest percentage, but the lowest number of total cases.


## Visualisation 4
> Test performend by diffenent countries to contain the growth  

Below, we see the testing results from each country. The "percent positive" is the percent of tests that turned out to be positive, out of every test conducted.  

Note that each line begins when the country begins to record/release their testing statistics.





```
def test_ratio(row):
  if row['total_tests']==0:
    return 0
  if row['total_cases'] / row['total_tests'] == 1:
    return 0
  return row['total_cases'] / row['total_tests']
```

```
#population data
from plotly.subplots import make_subplots
pop = pd.read_csv('https://raw.githubusercontent.com/eng22/misc/master/Populations.csv')
#data.plot(y=['total_cases', 'total_tests'], use_index=True, figsize=(10,10))

test = pd.read_csv('https://raw.githubusercontent.com/eng22/misc/master/owo.csv')

#Create percentage of positive tests in dataframe
test['Percent Positive Tests'] = test.apply (lambda row: test_ratio(row), axis=1)

fig = make_subplots(rows=1, cols=1)


#fig = px.line(test.loc[test['iso_code']=='DNK'], x='date', y='Percent Positive Tests')


fig.add_scatter(x=(test.loc[test['iso_code']=='DNK'])['date'],
                y=(test.loc[test['iso_code']=='DNK'])['Percent Positive Tests'],
                name='Denmark', text='Denmark')

fig.add_scatter(x=(test.loc[test['iso_code']=='NOR'])['date'],
                y=(test.loc[test['iso_code']=='NOR'])['Percent Positive Tests'],
                name='Norway', text='Norway')

fig.add_scatter(x=(test.loc[test['iso_code']=='ISL'])['date'],
                y=(test.loc[test['iso_code']=='ISL'])['Percent Positive Tests'],
                name='Iceland', text='Iceland')


fig.add_scatter(x=(test.loc[test['iso_code']=='SWE'])['date'],
                y=(test.loc[test['iso_code']=='SWE'])['Percent Positive Tests'],
                name='Sweden', text='Sweden')

fig.add_scatter(x=(test.loc[test['iso_code']=='FIN'])['date'],
                y=(test.loc[test['iso_code']=='FIN'])['Percent Positive Tests'],
                name='Finland', text='Finland')

fig.update_layout(
    title_text = 'Percent of Positive COVID-19 Tests'
)


fig.show()
```

#**4. Genre**
**Which genre of data story was used?**  
The seven basic genres in narrative visualization are magazine-style, annotated chart, partitioned poster, flow chart, comic strip, slide show, and film/video/animation. 

Such genre varies mainly from the point of view of:


> Distinct visual scenes, multiplexed in time and/or space.

> The ordering of their visual elements.




However, these categories are not generally exclusive: they may work in the same way as key components, merging to create more diverse visual styles. All these genres perform for a specific task for the specific narrate type.  Picking the appropriate genre rely on several factors, including the type of story, type of dataset, target audiences, the sophistication of the plot. 

We considered ***'Annotated Graphs/maps'*** to address the visualization since we are showcasing the comparison of the response to COVID-19 between Scandinavian countries, it will be convenient to outcast the features like multi-messaging where it can communicate clearly through the interaction of the text with the annotations and graphic elements. Users can discover additional details as per the requirements by hovering highlights and use other tools. This also retains a consistent visual platform that only alters content within each panel while leaving the general layout of the visual elements unchanged. There are several other genres that has been covered along with the annotated graphs.[2]
